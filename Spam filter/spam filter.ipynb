```python
import pandas as pd

data = pd.read_csv("spam.csv")
print(data)

bow = {}
mails = data["MAIL"]
tot_word_count = 0

for mail in mails:
    mail = mail.split()
    for word in mail:
        if word in bow:
            bow[word] += 1
            tot_word_count += 1
        else:
            bow[word] = 1
            tot_word_count += 1
            
print(bow)

fil = {'spam': 0, 'ham': 0}
for fils in data["fil"]:
    fil[fils] += 1
print(fil)
print("Total words: ", tot_word_count)

prob = {'spam': {}, 'ham': {}}
for index, row in data.iterrows():
    mail = row['MAIL']
    res = row['fil']
    done = []
    for word in mail.split():
        if word not in done:
            done.append(word)
            if word not in prob[res]:
                prob[res][word] = 1
            else:
                prob[res][word] += 1
target_prob = {'spam': fil['spam']/len(data), 'ham': fil['ham']/len(data)}

for res in prob:
    for word in prob[res]:
        prob[res][word] = prob[res][word] / fil[res]
print(prob)
print(target_prob)

word = "send"

p_ws = prob['spam'].get(word, 0)
p_s = target_prob['spam']
p_wh = prob['ham'].get(word, 0)
p_h = target_prob['ham']

p_sw = p_ws * p_s / (p_ws * p_s + p_wh * p_h)
print(p_sw)

word1 = "review"
word2 = "password"

p_w1s = prob['spam'].get(word1, 0)
p_s = target_prob['spam']
p_w1h = prob['ham'].get(word1, 0)
p_h = target_prob['ham']

p_w2s = prob['spam'].get(word2, 0)
p_w2h = prob['ham'].get(word2, 0)

p_sww = p_w1s * p_w2s * p_s / (p_w1s * p_w2s * p_s + p_w1h * p_w2h * p_h)
print(p_sww)
```
